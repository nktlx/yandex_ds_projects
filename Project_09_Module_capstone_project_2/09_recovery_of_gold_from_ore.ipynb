<h3> Описание данных <a class="tocSkip"> </h3>

<h4> Технологический процесс: <a class="tocSkip"> </h4>
    
- Rougher feed — исходное сырье
- Rougher additions (или reagent additions) — флотационные реагенты: Xanthate, Sulphate, Depressant
 - Xanthate — ксантогенат (промотер, или активатор флотации);
 - Sulphate — сульфат (на данном производстве сульфид натрия);
 - Depressant — депрессант (силикат натрия).
- Rougher process (англ. «грубый процесс») — флотация
- Rougher tails — отвальные хвосты
- Float banks — флотационная установка
- Cleaner process — очистка
- Rougher Au — черновой концентрат золота
- Final Au — финальный концентрат золота

<h4>Параметры этапов: <a class="tocSkip"> </h4>

- air amount — объём воздуха
- fluid levels — уровень жидкости
- feed size — размер гранул сырья
- feed rate — скорость подачи

<h4>Наименование признаков: <a class="tocSkip"> </h4>
    
- Наименование признаков должно быть такое: `[этап].[тип_параметра].[название_параметра]`
- Пример: `rougher.input.feed_ag`
- Возможные значения для блока `[этап]`:
 - `rougher` — флотация
 - `primary_cleaner` — первичная очистка
 - `secondary_cleaner` — вторичная очистка
 - `final` — финальные характеристики
- Возможные значения для блока `[тип_параметра]`:
 - `input` — параметры сырья
 - `output` — параметры продукта
 - `state` — параметры, характеризующие текущее состояние этапа
 - `calculation` — расчётные характеристики
 
<h4>Целевые признаки:  <a class="tocSkip"> </h4>
    
- `rougher.output.recovery` - эффективность обогащения чернового концентрата;
- `final.output.recovery` - эффективность обогащения финального концентрата.

### Проверка расчета эффективности обогащения 

Проверим, что признак `rougher.output.recovery` подсчитан верно на обучающей выборке. Для начала рассчитаем вручную значение этого признака и запишем в отдельной переменной.

```python
C = df_train['rougher.output.concentrate_au'] # доля золота в концентрате после флотации
F = df_train['rougher.input.feed_au'] # доля золота в сырье до флотации
T = df_train['rougher.output.tail_au'] # доля золота в отвальных хвостах после флотации

recovery = (C * (F - T)) / (F * (C - T)) * 100
recovery.head()
```

Сравним полученные значения с теми, что находятся в нашей таблице и найдём *MAE* между ними.

```python
pd.concat([df_train['rougher.output.recovery'], recovery], axis=1).head(10)
```

```python
mae = mean_absolute_error(df_train['rougher.output.recovery'], recovery)

print('MAE между полученными расчётами и табличными данными: ', mae)
```

Метрика *MAE* достаточно низкая, чтобы говорить о том, что между значениями признакамов нет значительной разницы. В таблице указана верная эффективность обогащения чернового концентрата.

## Анализ данных

### Изменение концентрации металлов

Посмотрим, как меняется концентрация металлов (Au, Ag, Pb) на различных этапах: в сырье, в черновом концентрате, в концентрате после первой очистки и в финальном концентрате.

```python
# функция построения распределения концентрации метлла
def plot_metal_distribution(data, metal):
    raw = data[f'rougher.input.feed_{metal}']
    rough = data[f'rougher.output.concentrate_{metal}']
    clean = data[f'primary_cleaner.output.concentrate_{metal}']
    final = data[f'final.output.concentrate_{metal}']
    fig, ax = plt.subplots(figsize=(15, 5))
    ax.hist(raw, bins=100, alpha=0.7)
    ax.hist(rough, bins=100, alpha=0.7)
    ax.hist(clean, bins=100, alpha=0.7)
    ax.hist(final, bins=100, alpha=0.7)
    ax.legend(['в сырье', 'в черновом концетрате', 'после первой очистки', 'в финальном концентрате'])
    ax.set_xlabel(f'Концентрация {metal}')
    ax.set_ylabel('Количество объектов')
    ax.set_title(f'Распределение концентрации {metal} на разных стадиях')
    plt.show()
```

```python
plot_metal_distribution(df_full,'au')
```

Судя по графикам, концентрация золота в обрабатываемом материале возрастает после каждого из этапов очистки. Мы можем увдеть, что большая часть объектов после финальной обработки содержит концентрацию золота в значениях от **40** до **50**.

```python
plot_metal_distribution(df_full, 'ag')
```

Концентрация серебра после флотации возрастает, а затем в следствии последующих этапов очистки снижается. После финального этапа концентрация серебра в большинстве объектов даже ниже, чем до флотации.

```python
plot_metal_distribution(df_full, 'pb')
```

Концентрация свинца в веществе после каждого этапа очистки повышается, однако после второй очистки рост концентрации не такой значительный, как после предыдущих этапов. Стоит отметить, что на всех трех графиках присутствует некоторое количество объектов, которые содержат нулевую концентрацию рассматриваемого металла на различных этапах обработки.

Построим таблицу изменения концентраций металлов, чтобы убедиться в наших рассуждениях на числах. В качестве значений таблицы используем средние показатели концентрации для всех объектов.

```python
metals = ['au', 'ag', 'pb']
avg_values = []

for metal in metals:
    raw = df_full[f'rougher.input.feed_{metal}'].mean()
    rough = df_full[f'rougher.output.concentrate_{metal}'].mean()
    clean = df_full[f'primary_cleaner.output.concentrate_{metal}'].mean()
    final = df_full[f'final.output.concentrate_{metal}'].mean()
    avg_values.append((raw, rough, clean, final))
    
avg_concentrates = pd.DataFrame(avg_values, columns=['raw', 'rough', 'clean', 'final'], index=metals)
avg_concentrates
```

Полученные данные соответствует описанной ранее ситуации. Концентрация средняя `au` и `pb` в объектах возрастает после каждого этапа очистки, а концентрация `ag` сначала возрастаёт, а затем опускается даже ниже исходного значения.

Также можно увидеть некоторое количество объектов с нулевой концентрацией ценных металлов на разных этапах. Посмотрим, сколько таких объектов. Рассматривать будем `df_train`, поскольку нам необходимо иметь корректные обучающие данные для дальнейшего построения моделей.

```python
metal_concetrate = []

for metal in metals:
    metal_concetrate.append(df_train[f'rougher.input.feed_{metal}'])
    metal_concetrate.append(df_train[f'rougher.output.concentrate_{metal}'])
    metal_concetrate.append(df_train[f'primary_cleaner.output.concentrate_{metal}'])
    metal_concetrate.append(df_train[f'final.output.concentrate_{metal}'])

# вспомогательный датафрейм с концентрацией металлов на каждом этапе
df_concentrates = pd.concat(metal_concetrate, axis=1)
df_concentrates.head()
```

```python
df_concentrates.eq(0).sum()
```

С первого взгляда может показаться, что нулей в каждом признаке слишком много, но мы можем заметить, что количество нулей на каждом этапе очистки у каждого металла одно и то же. Скорее всего это одни и те же объекты, в которых не меняется концентрация металлов после каждого этапа очистки. Удалим объекты с нулевыми значениями концентрации и вычислим количество удаленных строк.

```python
size_before = df_train.shape[0]
print('Текущий размер выборки', size_before)
```

```python
# удаляем нулевые строки из обучающей выборки
for column in df_concentrates.columns:
    df_train = df_train[df_train[column] != 0]

size_after = df_train.shape[0]
print('Размер выборки после удаления нулей: ', size_after)
```

Наши предположения подтвердились. В обучающей выборке **591** объект, который имеет нулевую суммарную концентрацию металлов на каком-либо из этапов очистки. Это довольно небольшое количество, допустимое для удаления. Эти объекты являются аномальными, а нам важно для обучения использовать корректные данные, поэтому эти строки стоить исключить.

```python
round((1 - size_after/size_before) * 100, 2)
```

Из тренировочного датафрейма мы в результате удалили около **4%** строк. Это приемлемо.

## Модель

Подготовим выборки  к машинному обучению. Из тренировочной выборки нам нужно извлечь целевые признаки, а для тестовой выборки нам нужно получить значения целевых признаков с помощью исходного датафрейма. Кроме того, из тренировочной выборки нужно исключить признаки, которых нет в тестовой.

```python
features_train = df_train[df_test.columns]
target_train = df_train[['rougher.output.recovery','final.output.recovery']]
features_test = df_test
target_test = df_full[['rougher.output.recovery','final.output.recovery']].loc[df_test.index]

print('Итоговые размеры выборок')
print('Признаки для обучения: ', features_train.shape)
print('Целевые признаки для обучения: ', target_train.shape)
print('Признаки для теста: ',features_test.shape)
print('Целевые признаки для теста: ',target_test.shape)
```

###  Функция итоговой sMAPE

Напишием функцию для вычисления итоговой sMAPE.

```python
def smape(predictions, target):
    numerator = np.abs(target - predictions)
    denominator = (np.abs(target) + np.abs(predictions)) / 2
    return np.mean(numerator / denominator) * 100
```

```python
def final_smape(rougher, final):
    return 0.25 * rougher + 0.75 * final
```

Поскольку мы создали собственную функцию для оценки качества модели, то нам для использования кросс-валидации нужно еще добавить собственную скор-функцию, совместимую с sklearn. Сделаем это, передав функцию вычисляющую sMAPE в *make_scorer()*. Из-за того, что в нашей задаче мы пытаемся минимизировать целевую метрику, данная функция будет возвращать отрицальтельное значение метрики, но мы будем оценивать абсолютное значение.

```python
smape_scorer = make_scorer(smape, greater_is_better=False)
```

Функции для вычисления метрик готовы. Переходим к обучению.

### Обучение и оценка качества моделей

Нам необходимо предсказать два целевых признака: эффективность обогащения чернового концентрата `rougher.output.recovery` и эффективность обогащения финального концентрата `final.output.recovery`. Для начала разебремся с первым.

Неправильно будет обучать модель на признаках, которые мы получаем в результате следующих этапов очистки, поскольку в ходе процесса нам недоступны данные из будущего.

```python
# извлекаем только доступные перед флотацией признаки, их должно быть в сумме 22
features_train_rougher = features_train.loc[:,'rougher.input.feed_au':'rougher.input.floatbank11_xanthate']
features_test_rougher = features_test.loc[:,'rougher.input.feed_au':'rougher.input.floatbank11_xanthate']
# подготавливаем целевой признак
target_train_rougher = target_train['rougher.output.recovery']
target_test_rougher = target_test['rougher.output.recovery']

display(features_train_rougher.shape)
display(target_train_rougher.shape)
```
