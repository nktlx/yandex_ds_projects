# Исследование успешности видеоигр


Из открытых источников  взяты исторические данные о продажах игр, оценки пользователей и экспертов, жанры и платформы (например, Xbox или PlayStation). Необходимо выявить определяющие успешность игры закономерности. Это позволит сделать ставку на потенциально популярный продукт и спланировать рекламные кампании. Представим, что нам необходимо сделать прогноз на следующий год.

В наборе данных попадается аббревиатура ESRB (Entertainment Software Rating Board) — это ассоциация, определяющая возрастной рейтинг компьютерных игр. ESRB оценивает игровой контент и присваивает ему подходящую возрастную категорию, например, «Для взрослых», «Для детей младшего возраста» или «Для подростков».

**Описание данных**

- `Name` — название игры
- `Platform` — платформа
- `Year_of_Release` — год выпуска
- `Genre` — жанр игры
- `NA_sales` — продажи в Северной Америке (миллионы проданных копий)
- `EU_sales` — продажи в Европе (миллионы проданных копий)
- `JP_sales` — продажи в Японии (миллионы проданных копий)
- `Other_sales` — продажи в других странах (миллионы проданных копий)
- `Critic_Score` — оценка критиков (максимум 100)
- `User_Score` — оценка пользователей (максимум 10)
- `Rating` — рейтинг от организации ESRB (англ. Entertainment Software Rating Board). Эта ассоциация определяет рейтинг компьютерных игр и присваивает им подходящую возрастную категорию.

**Цель исследования** 


Провести анализ и выяснить:
1. Cколько игр выпускалось в разные годы?
2. Как менялись продажи по платформам?
3. Какие платформы лидируют по продажам, растут или падают?
4. Как влияют на продажи внутри одной популярной платформы отзывы пользователей и критиков?
5. Выделяются ли жанры с высокими и низкими продажами?

По итогам анализа необходимо:
1. Составить портрет пользователя каждого региона. 
2. Проверить гипотезу о том, что средние пользовательские рейтинги платформ Xbox One и PC одинаковые.
3. Проверить гипотезу о том, средние пользовательские рейтинги жанров Action и Sports разные.

**Ход исследования**

Данные о с информацией о клиентах получены из файла `games.csv`. О качестве данных ничего не известно. Поэтому перед началом анализа понадобится обзор данных.

Затем, можно переходить к исследовательскому анализу, благодаря которому удасться получить необходимые для формирования будущей рекламной компании результаты. Другая часть — это данные о том, как игра была воспринята сообществом, например, пользовательская оценка или количество продаж в определенном регионе.

Преварительно можно утверждать, что данных достаточно для ответа на поставленные вопросы, но может потребоваться добавление новых столбцов на основе имеющихся.

В некоторых стобцах содержаться пропуски, где-то пропусков даже больше половины от общего числа строк. Явных дубликатов на этом этапе не обнаружено. Чтобы двигаться дальше, нужно устранить проблемы в данных.

### Предобработка данных

#### Изименение названий столбцов

В названиях столбцов используется не *snake_case*, это необходимо исправить.

````python
data.columns = data.columns.str.lower()
data.columns
````

#### Обработка пропусков

Выведем график для оценки масштаба проблемы с пропусками.

````python
# check
# пропущенные значения бары

def pass_value_barh(df):
    try:
        (
            (df.isna().mean()*100)
            .to_frame()
            .rename(columns = {0:'space'})
            .query('space > 0')
            .sort_values(by = 'space', ascending = True)
            .plot(kind = 'bar', figsize = (12 ,5), rot=0, legend = False, grid = True, 
                  xlabel='Столбцы', ylabel='Количество пропусков в %')
            .set_title('График оценки количества пропусков', fontsize = 15)    
        );    
    except:
        print('пропусков не осталось :) или произошла ошибка в первой части функции ')
````

````python
pass_value_barh(data)
````

На данный момент датафрейм содержит **16715** строк. Количество отбрасываемых данных не должно превышать 10% от изначального объема, поэтому, обрабатывая пропуски, мы не можем удалить более **1670** строк.

````python
data.isna().sum()
````

````python
data[data['year_of_release'].isna()]
````

Cтолбец `year_of_release` содержит позволительное для удаления количество пропусков, всего **269**. Устраним их.

````python
data = data.dropna(subset=['year_of_release'])
data['year_of_release'].isna().sum()
````

````python
data[(data['name'].isna()) | (data['genre'].isna())]
````

Столбцы `name` и `genre` содержат по **2** пропуска. Поскольку дальнейший анализ объектов с пропусками в этих двух параметрах бессмысленен, и количество пропущенных значений крайне мало, то просто удалим их.

````python
data = data.dropna(subset=['name', 'genre'])
data[['name', 'genre']].isna().sum()
````

Рассмотрим столбец `user_score`. Он может хранить неожиданные значения, поскольку это текстовый столбец и информацию о нём не удалось получить с помощью метода describe.

````python
data['user_score'].unique()
````

В числе уникальных объектов обнаруживаем значение **tbd**, которое в данном контексте обозначает то же, что и **nan**. Приведём все подобные строки к одному типу.

````python
data.loc[data['user_score'] == 'tbd', 'user_score'] = np.nan
data['user_score'].isna().sum()
````

В столбцe `user_score`, как и в столбцах `critic_score`, `rating` крайне большое количество пропусков. В данном случае не стоит пытаться заполнить пропуски с помощью средних/медианных значений или на основе информации из других столбцов, так как это может исказить общую "картину" и привести к некорректному результату при анализе. Просто удалить их тоже нельзя, мы можем потерять более половины от всего датафрейма.

Посмотрим сколько строк имеют пропуски во всех трех столбцах.

````python
temp_data = data[(data['critic_score'].isna()) & 
                 (data['user_score'].isna()) & 
                 (data['rating'].isna())]
temp_data
````

Пропуски одновременно во всех трех столбцах встречаются в **6580** случаях. Это все еще то значение, которым мы не можем пожертвовать, однако в данных можно обнаружить строки, где помимо пропусков, практически нулевые показатели по продажам в регионах.

````python
temp_data = temp_data[(temp_data['na_sales'] < 0.02) & 
                      (temp_data['eu_sales'] < 0.02) & 
                      (temp_data['jp_sales'] < 0.02) & 
                      (temp_data['other_sales'] < 0.02)]
temp_data
````

В выборке **491** игра, для которой при очень низких продажах отсуствует важная информация. Это уже допустимое число строк, от которых можно просто избавиться. Такие данные не представляют практически никакой ценности для дальнейшего анализа, но могут оказывать нежелательное влияние на средние значения. Сократим количество пропусков, исключив данные строки из выборки.

````python
data = data.drop(temp_data.index)
len(data)
````

Посмтрим на уникальные значения в столбце `rating`. 

````python
data['rating'].unique()
````

Рейтинг "K-A" ("Kids to Adults") обозначает, что игра подоходит для всех возрастов. На сегодняшний день играм этой категории присваивают рейтинг "E"("Everyone"), а "K-A" — это устаревшее обозначение. Поскольку эти рейтинги обозначают одно и то же, то мы можем привести все строки с подобными занчениями к единому варианту.

````python
data['rating'].replace('K-A','E', inplace=True)
data['rating'].unique()
````

````python
data[data['rating'].isna()]
````

Поскольку рейтинг — категориальные данные, то заполним пропуски в данном столбце значением "unknown".

````python
data['rating'] = data['rating'].fillna('unknown')
data['rating'].isna().sum()
````

#### Изменение типов данных

В некоторых столбцах тип данных не соотвуствует значениям, которые в них хранятся. Это может вызвать затруднения при обработке и анализе данных, поэтому параметры нужно привести к верному типу.

Приведед значения в столбце `year_of_release` в целочисленному типу, поскольку год не может быть дробным. Это улучшит восприятия данных и сэкономит память.

````python
data['year_of_release'] = data['year_of_release'].astype('int32')
data['year_of_release'].dtype
````

Переведем данные в столбце `user_score` из строкового формата в числовой, так как пользовательская оценка выражается числом.

````python
data['user_score'] = pd.to_numeric(data['user_score'])
data['user_score'].dtype
````

````python
data.info()
````

#### Подсчет суммарных продаж во всех регионах

Добавим в датафрейм новый столбец `total_sales`, в котором будем хранить суммарное количество продаж во всех реегионах.

````python
data['total_sales'] = data[['na_sales','eu_sales','jp_sales', 'other_sales']].sum(axis = 1)
data.head(15)
````

#### Поиск неявных дубликатов

Явных дубликатов проекте обнаружено не было, попробуем найти и устранить неявные дубликаты.

Передведём все значения в столбцах строкового типа к единому регистру.

````python
columns_to_lower = ['name', 'platform', 'genre']
for column in columns_to_lower:
    data[column] = data[column].str.lower() 
        
data.head()
````

Найдем строки, в которых совпадает название игры и название платформы, для которой она выпускалась.

````python
data[data[['name', 'platform']].duplicated(keep=False)]
````

Для игры "madden nfl 13" нашелся очевидный дубликат, его стоит изключить из датафрейма. Для игры "need for speed" не всё так просто. На мой взгляд, не стоит исключать данные дубликаты из датафрейма, поскольку даты выхода игр различаются. В 2012 году произошел полноценный перевыпуск данный игры для "PC" и "X360" с обновленной графикой сюжетом и управлением. Вышедшие в один год и тот же год релизы тоже нельзя считать дубликатами, поскольку они были выпущены для разных платформ. Удаляем только дубликат с индексом **16230**.

````python
data = data.drop(index=16230)
len(data)
````

````python
# check
data.info()
````

**Вывод**

Предобработка обнаружила три проблемы в данных:
- неверный регистр в названии столбцов,
- пропущенные значения,
- неверные типы данных,
- неявные дубликаты.

Также мы преобразовали типы данных в некоторых столбцах так, чтобы облегчить работу с ними. Неявные дубликаты были исключены.

### Исследовательский анализ данных

#### Первичный анализ

Выясним, сколько игр выпускалось в разные годы.

````python
years_pivot = data.pivot_table(index='year_of_release', values='name', aggfunc='count')
years_pivot
````

````python
years_pivot.plot(kind ='bar', figsize = (12, 5))
plt.title('Количество выпущенных игр по годам')
plt.xlabel('Год')
plt.ylabel('Количество вышедших игр')
plt.show()
````

Судя по графику, можно сказать, что основная масса игр была выпущена в период с 2001 по 2016 года. Больше всего игр было выпущено в 2008 и 2009 года — около **1400** штук в каждый из них. Данные по играм до 1994 года включительно можно не считать важными, поскольку в те годы выходило гораздо меньше игр, чем в последующие годы.

Посмотрим, как менялись продажи по платформам. Выберем платформы с наибольшими суммарными продажами и построим распределение по годам.

````python
platform_pivot = (data
                  .pivot_table(index='platform', values='total_sales', aggfunc='sum')
                  .sort_values(by='total_sales', ascending=False)
                  .head(10))
platform_pivot
````

````python
platform_pivot.plot(kind ='bar', figsize = (10, 5), rot = 0)
plt.title('Платформы с самыми высокими показателями по продажам')
plt.xlabel('Платформа')
plt.ylabel('Суммарное количество продаж (млн. копий)')
plt.show()
````

Таблица и столбчатая диаграмма отражают топ-10 платформ с наибольними проджами. Так абсолютным лидером по продажам видеоигр является платформа "PS2", для которой было продано более **1.2** млрд. копий. 

Можно увидеть явную разницу в продажах игр для платформ "PS" и "PS4". Консоль "PS" более чем в 2 раза превосходит по продажам следующую строчку данного рейтинга. На мой взгляд, следует ограничить рейтинг лидеров, оставив в нём лишь первые 6 позиций, то есть до "PS" включительно.

````python
platform_pivot = platform_pivot.head(6)
platform_pivot
````

В топе остались лишь явные лидеры. Далее можно переходить к пострению распределения продаж видеоигр по годам.

````python
# считаем продажи по годам для каждой платформы
platform_year_pivot = data.pivot_table(index=['platform','year_of_release'], 
                                       values='total_sales', 
                                       aggfunc='sum')
platform_year_pivot
````

````python
# для выделенных ранее платформ строим диаграммы
for platform in platform_pivot.index:
    platform_year_pivot.loc[platform].plot(kind = 'bar', figsize = (7, 4), rot=0)
    plt.title('Изменение количества продаж по годам для ' + platform )
    plt.xlabel('Год')
    plt.ylabel('Продажи (млн. копий)')
    plt.show()
````

Характерным сроком, за который появляются новые и исчезают старые платформы можно назвать примерно 8 лет.

Можно сказать, что именно в течении этого срока платформы в нашем топе остаются востребованными, а затем они теряют свою актуальность. ### Проверка гипотез

Проверим гипотезу, что средние пользовательские рейтинги платформ "Xbox One" и "PC" одинаковые.

```
H_0: Пользовательский рейтинг (user_score) для платформы "Xbox One" = пользовательский рейтинг (user_score) для платформы "PC"`
H_1: Пользовательский рейтинг (user_score) для платформы "Xbox One" ≠ пользовательский рейтинг (user_score) для платформы "PC"`
alpha = 0.05
```

````python
avg_score_XOne = actual_data[actual_data['platform'] == 'xone']['user_score'].dropna()
avg_score_PC = actual_data[actual_data['platform'] == 'pc']['user_score'].dropna()
# results = вызов метода для проверки гипотезы
results = st.ttest_ind(avg_score_XOne, avg_score_PC, equal_var = False)
# alpha = значение уровня значимости
alpha = 0.05
print(results.pvalue)
if results.pvalue < alpha:
    print('Отвергаем нулевую гипотезу ')
else:
    print('Не получилось отвергнуть нулевую гипотезу')
````

Судя по значению p-value в **0.15**, средние оценки пользователей для игр на платформах "Xbox One" и "PC" могут быть равны. Гипотеза не отвергнута.

Проверим гипотезу, что средние пользовательские рейтинги жанров "Action" и "Sports" разные.

```
H_0: Пользовательский рейтинг (user_score) для жанра "Action" = пользовательский рейтинг (user_score) для жанра "Sports"`
H_1: Пользовательский рейтинг (user_score) для жанра "Action" ≠ пользовательский рейтинг (user_score) для жанра "Sports"`
alpha = 0.05
```

````python
avg_score_action = actual_data[actual_data['genre'] == 'action']['user_score'].dropna()
avg_score_sports = actual_data[actual_data['genre'] == 'sports']['user_score'].dropna()
# results = вызов метода для проверки гипотезы
results = st.ttest_ind(avg_score_action, avg_score_sports, equal_var = False)
# alpha = значение уровня значимости
alpha = 0.05
print(results.pvalue)
if results.pvalue < alpha:
    print('Отвергаем нулевую гипотезу ')
else:
    print('Не получилось отвергнуть нулевую гипотезу')
````

Судя по значению p-value в **1.45e-20**, средние оценки пользователей на игры в жанре "Action" и на игры в жанре "Sports" не равны. Мы отвергаем гипотезу H_0 в пользу гипотезы H_1.

**Вывод**

В данном разделе была произведена проверка следующих гипотез: 
- средние пользовательские рейтинги платформ *Xbox One* и *PC* одинаковые,
- средние пользовательские рейтинги жанров *Action* и *Sports* разные.
